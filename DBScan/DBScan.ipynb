```python
import pandas as pd

# Loading the dataset
file_path = '/Users/aviajmera/Downloads/Clustering/DBScan/Country-data.csv'
data = pd.read_csv(file_path)

# Displaying the first few rows of the dataset to understand its structure
data.head()
```

```python
!pip install pycaret
from pycaret.clustering import *

# Setting up the PyCaret environment
# Ignoring the 'country' column as it's a categorical variable and not useful for clustering in this context
cluster_setup = setup(data = data.drop('country', axis=1), silent=True, session_id=123)

# Creating a DBSCAN model
dbscan_model = create_model('dbscan', eps=0.5, min_samples=5)

# Assigning the cluster labels to the dataset
dbscan_results = assign_model(dbscan_model)

# Displaying the first few rows of the dataset with cluster labels
dbscan_results.head()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Preprocessing the data (excluding the 'country' column)
X = data.drop('country', axis=1)
X_scaled = StandardScaler().fit_transform(X)

# Performing DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)

# Adding the cluster labels to the original dataset
data_with_clusters = data.copy()
data_with_clusters['Cluster'] = clusters

# Displaying the first few rows of the dataset with cluster labels
data_with_clusters.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Counting the number of countries in each cluster
cluster_counts = data_with_clusters['Cluster'].value_counts().sort_index()

# Plotting the distribution of countries in each cluster
plt.figure(figsize=(10, 6))
sns.barplot(x=cluster_counts.index, y=cluster_counts.values)
plt.title('Distribution of Countries in Clusters (DBSCAN)')
plt.xlabel('Cluster Label')
plt.ylabel('Number of Countries')
plt.xticks(range(len(cluster_counts)), cluster_counts.index)
plt.show()

# Displaying the counts for each cluster
cluster_counts
```

The distribution of countries across the clusters reveals that all 167 countries in the dataset are labeled as noise (Cluster = -1) with the current DBSCAN parameters (eps=0.5, min_samples=5). This indicates that the chosen parameters are too restrictive, and no dense regions are identified in the dataset.

```python
# Adjusting the parameters for DBSCAN
# Trying a larger value for eps and a smaller value for min_samples
dbscan_adjusted = DBSCAN(eps=2, min_samples=3)
clusters_adjusted = dbscan_adjusted.fit_predict(X_scaled)

# Adding the adjusted cluster labels to the dataset
data_with_clusters['Cluster_Adjusted'] = clusters_adjusted

# Counting the number of countries in each adjusted cluster
adjusted_cluster_counts = data_with_clusters['Cluster_Adjusted'].value_counts().sort_index()

# Plotting the distribution of countries in each adjusted cluster
plt.figure(figsize=(10, 6))
sns.barplot(x=adjusted_cluster_counts.index, y=adjusted_cluster_counts.values)
plt.title('Distribution of Countries in Adjusted Clusters (DBSCAN)')
plt.xlabel('Adjusted Cluster Label')
plt.ylabel('Number of Countries')
plt.xticks(range(len(adjusted_cluster_counts)), adjusted_cluster_counts.index)
plt.show()

# Displaying the counts for each adjusted cluster
adjusted_cluster_counts, data_with_clusters.head()
```

```python
After adjusting the parameters of the DBSCAN algorithm (eps=2, min_samples=3), we have a more meaningful clustering of the countries. The distribution of countries across the clusters is now as follows:

Cluster 0: Contains 156 countries. This cluster represents the majority of the countries in the dataset.
Cluster -1: Contains 11 countries. These are considered noise or outliers according to the DBSCAN algorithm. This could mean that these countries have unique characteristics that significantly differ from the rest of the dataset.
The first five rows of the dataset, now including the adjusted cluster labels, provide a glimpse into how countries are categorized. For instance, Afghanistan, Albania, Algeria, Angola, and Antigua and Barbuda are all part of Cluster 0.
```
