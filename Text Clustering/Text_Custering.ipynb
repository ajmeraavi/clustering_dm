```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import AgglomerativeClustering
import numpy as np
```

```python
embedder = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
corpus = [
    "The moon cast a soft glow on the tranquil lake.",
    "Artificial intelligence is revolutionizing various industries.",
    "Exploring underwater caves is an adventurous and challenging activity.",
    "Poetry captures the beauty of emotions in words.",
    "Innovation is the key to progress in the rapidly changing world.",
    "Birdwatching requires patience and a keen eye for detail.",
    "The city skyline at night is a breathtaking sight.",
    "Sailing on the open sea brings a sense of freedom and serenity.",
    "Robotics is advancing rapidly, transforming how we live and work.",
    "Hiking in the mountains offers stunning vistas and a sense of accomplishment.",
    "The laughter of children is a source of pure joy.",
    "Virtual reality technology is reshaping the way we experience entertainment.",
    "Baking a perfect soufflé demands precision and skill.",
    "Space exploration inspires awe and fuels scientific discovery.",
    "The laughter of friends gathered around a bonfire is unforgettable.",
    "Time management is crucial for achieving personal and professional goals.",
    "Dance is a universal language that expresses emotions without words.",
    "The aroma of freshly brewed coffee is invigorating.",
    "Environmental sustainability is vital for preserving our planet.",
    "The art of storytelling has been a part of human culture for centuries.",
    "Mindfulness meditation promotes inner peace and self-awareness.",
    "Architectural marvels tell the story of human ingenuity.",
    "Astronauts experience the beauty of Earth from a unique perspective.",
    "Solving puzzles stimulates cognitive abilities and creativity.",
    "The symphony of nature can be heard in the rustling leaves and chirping birds.",
    "Mind-bending illusions challenge our perception of reality.",
    "Exploring ancient ruins provides a glimpse into civilizations of the past.",
    "A cup of hot tea is comforting on a cold winter's day.",
    "Stargazing on a clear night sparks a sense of wonder and awe.",
]
```

```python
corpus_embeddings = embedder.encode(corpus)
corpus_embeddings = corpus_embeddings /  np.linalg.norm(corpus_embeddings, axis=1, keepdims=True)
corpus_embeddings
```

```python
clustering_model = AgglomerativeClustering(n_clusters=None, distance_threshold=1.5) 
clustering_model.fit(corpus_embeddings)
cluster_assignment = clustering_model.labels_
```

```python
clustered_sentences = {}
for sentence_id, cluster_id in enumerate(cluster_assignment):
    if cluster_id not in clustered_sentences:
        clustered_sentences[cluster_id] = []

    clustered_sentences[cluster_id].append(corpus[sentence_id])
```

```python
for i, cluster in clustered_sentences.items():
    print("Cluster ", i+1)
    print(cluster)
    print("")
```
